```python
import json

import pandas as pd
```

```python
ml = pd.read_parquet('~/repos/zoobot/gz_decals_auto_posteriors.parquet')
```

```python
vols = pd.read_parquet('~/repos/zoobot/dr5_volunteer_catalog_internal.parquet')
# TODO add DR2 as well, just need the urls
```

```python
ml_cols = ['iauname'] + [x for x in ml.columns if x.endswith('_fraction')] + [x for x in ml.columns if x.endswith('_asked')]
```

```python
vol_cols = ['iauname', 'locations']
```

```python
df = pd.merge(vols[vol_cols], ml[ml_cols], on='iauname', how='inner')
```

```python
df
```

```python
df['url'] = df['locations'].apply(lambda x: json.loads(x)[0])
```

```python
df['url']
```

```python
del df['locations']
```

```python
batch_size = 80000

current_index = 0
iterations = 0
while current_index < len(df):
    next_index = current_index + batch_size
    df[current_index:current_index+batch_size].to_csv('~/repos/galaxy-poster/decals_{}.csv'.format(iterations), index=False)
    iterations += 1
    current_index += batch_size
```
